```python
from typing import Any, Generator, NoReturn
import numpy as np

def wichmann_hill_rng(x0: int, y0: int, z0: int) -> Generator[float, Any, NoReturn]:
    multipliers = np.array([171, 172, 170])
    modulos = np.array([30269, 30307, 30323])
    X = np.array([x0, y0, z0]) % modulos
    while True:
        yield (X / modulos).sum() % 1
        X = (X * multipliers) % modulos
```

```python
from itertools import islice

N = int(4e3)
generator = wichmann_hill_rng(101, 431, 8085)
numbers = [x for x in islice(generator, N)]
numbers
```

```python
from matplotlib import pyplot as plt

for step in [1, 2]:
    xs = numbers[:-step:step]
    ys = numbers[step::step]
    plt.scatter(xs, ys, s=1)
    plt.title(f"Wichmann Hill RNG spatial structure for step of ${step}$")
    plt.xlabel("$k_n$")
    plt.ylabel(f"$k_{{n+{step}}}$")
    plt.show()
```
